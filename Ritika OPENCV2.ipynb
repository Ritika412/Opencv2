```python
##code for convert color image to blue image
import cv2
import numpy as np
import matplotlib.pyplot as plt

bgr_img = cv2.imread('C:/Users/pc/Pictures/bhai brtday.jpg')

b,g,r = cv2.split(bgr_img)       # get b,g,r
rgb_img = cv2.merge([r,g,b])     # switch it to rgb

plt.imshow(rgb_img)
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.show()
plt.imshow(rgb_img)
while True:
    k = cv2.waitKey(0) & 0xFF    # 0xFF? To get the lowest byte.
    if k == 27: break            # Code for the ESC key

cv2.destroyAllWindows()
```

```python
import cv2 
import matplotlib.pyplot as plt 

# Read an image 
img_bgr = cv2.imread('C:/Users/pc/Pictures/bhai brtday.jpg', 1) 
  
plt.imshow(img_bgr) 
plt.show() 
```

```python
# Histogram plotting of original image 
color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_bgr], 
                         [i], None, 
                         [256], 
                         [0, 256]) 
      
    plt.plot(histr, color = col) 
      
    # Limit X - axis to 256 
    plt.xlim([0, 256]) 
      
plt.show() 
```

```python
# Negate the original image 
img_neg = 1 - img_bgr 
  
plt.imshow(img_neg) 
plt.show() 
```

```python

# Histogram plotting of 
# negative transformed image 
color = ('b', 'g', 'r') 
  
for i, col in enumerate(color): 
      
    histr = cv2.calcHist([img_neg], 
                         [i], None,  
                         [256], 
                         [0, 256]) 
      
    plt.plot(histr, color = col) 
    plt.xlim([0, 256]) 
      
plt.show()
```

```python
##ans4
import cv2

face_cascade = cv2.CascadeClassifier('C:/Users/pc/Downloads/haarcascade_frontalface_default.xml')
# Read the image
img = cv2.imread('C:/Users/pc/Pictures/bhai brtday.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, 1.5, 3)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y), (x+w,y+h),(255,0,0),3)
    
cv2.imshow('img', img)
cv2.waitKey(0)
```
